```python
from pyspark.sql import functions as F

print("🔧 Executing schema migration to append HCAI lineage metadata...")

try:
    # 1. Read, transform, and overwrite studentinfo_bronze with schema evolution enabled
    print("Migrating schema for studentinfo_bronze...")
    df_info_raw = spark.read.table("studentinfo_bronze")
    df_info_repaired = df_info_raw.withColumn("PIPELINE_INGEST_TS", F.current_timestamp()) \
                                  .withColumn("SOURCE_FILE_NAME", F.lit("studentInfo.csv"))
    
    # Fix: Added overwriteSchema option to tell Delta Lake we are adding new columns
    df_info_repaired.write.mode("overwrite") \
                          .option("overwriteSchema", "true") \
                          .saveAsTable("studentinfo_bronze")
    print("✔ studentinfo_bronze schema successfully evolved!")

    # 2. Read, transform, and overwrite studentvle_bronze with schema evolution enabled
    print("Migrating schema for studentvle_bronze...")
    df_vle_raw = spark.read.table("studentvle_bronze")
    df_vle_repaired = df_vle_raw.withColumn("PIPELINE_INGEST_TS", F.current_timestamp()) \
                                .withColumn("SOURCE_FILE_NAME", F.lit("studentVle.csv"))
    
    # Fix: Added overwriteSchema option here as well
    df_vle_repaired.write.mode("overwrite") \
                         .option("overwriteSchema", "true") \
                         .saveAsTable("studentvle_bronze")
    print("✔ studentvle_bronze schema successfully evolved!")
    
    print("\n🚀 SUCCESS: OULAD tables successfully refactored and updated with metadata tags!")

except Exception as e:
    print(f"❌ SCHEMA EVOLUTION FAILURE: {str(e)}")
```
