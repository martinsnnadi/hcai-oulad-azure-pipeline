```python
import urllib.request
import zipfile
import os

# 1. Define the official OULAD download endpoint and storage target paths
SOURCE_URL = "https://analyse.kmi.open.ac.uk/open-dataset/download"
ZIP_LANDING_PATH = "/lakehouse/default/Files/oulad_raw.zip"
EXTRACT_TARGET_DIR = "/lakehouse/default/Files/raw_csv/"

print("[1/3] Initiating cloud-to-cloud file extraction stream...")
os.makedirs(EXTRACT_TARGET_DIR, exist_ok=True)

try:
    # Fetch the dataset directly into Fabric's persistent file system
    urllib.request.urlretrieve(SOURCE_URL, ZIP_LANDING_PATH)
    print(f"✔ Archive successfully fetched and landed at: {ZIP_LANDING_PATH}")
    
    # 2. Extract files on the cloud cluster allocation
    print("[2/3] Unzipping relational education data packages...")
    with zipfile.ZipFile(ZIP_LANDING_PATH, 'r') as zip_ref:
        zip_ref.extractall(EXTRACT_TARGET_DIR)
    print(f"✔ Extraction complete. Files isolated inside: {EXTRACT_TARGET_DIR}")
    
    # 3. Clean up the zip file to save cloud storage space
    os.remove(ZIP_LANDING_PATH)
    print("[3/3] Clean up complete. Workspace optimized.")
    
    print("\n🚀 SUCCESS: Verify the files listed below in your Bronze Workspace:")
    print(os.listdir(EXTRACT_TARGET_DIR))

except Exception as e:
    print(f"❌ INGESTION FAILURE ERROR: {str(e)}")
```
