```python
import os
import requests

# 1. Pipeline Environment Paths (Writes directly to Fabric's native file storage)
SOURCE_URL = "https://wwwn.cdc.gov/nchs/data/nhanes3/1a/adult.dat"
TARGET_FILE_PATH = "/lakehouse/default/Files/nhanes3/raw/adult.dat"

print("[1/2] Initiating memory-isolated HTTP network stream to CDC repository...")

try:
    # Ensure the destination directory exists within your Lakehouse Files section
    os.makedirs(os.path.dirname(TARGET_FILE_PATH), exist_ok=True)
    
    # Open connection using stable chunked streaming
    headers = {'User-Agent': 'Mozilla/5.0 DataEngineeringPipeline/1.0'}
    with requests.get(SOURCE_URL, headers=headers, stream=True, timeout=20) as r:
        r.raise_for_status()
        
        print(f"[2/2] Writing raw binary blocks straight to Lakehouse path: {TARGET_FILE_PATH}")
        # Stream raw text bytes directly to the file without loading the archive into RAM
        with open(TARGET_FILE_PATH, 'wb') as f:
            for chunk in r.iter_content(chunk_size=131072): # Efficient 128KB chunk buffers
                if chunk:
                    f.write(chunk)
                    
    print("\n🚀 SUCCESS: Flat file safely extracted as-is to your Fabric Lakehouse Files zone!")

except Exception as e:
    print(f"❌ FABRIC FLAT-FILE INGESTION FAILURE: {str(e)}")
    raise e
```
